# Eliminando datos

La función más común para eliminar datos es<br>
DataFrame.drop(labels=None, axis=0, index=None, columns=None)<br>
Parámetros
* labels: Etiqueta o índice de la fila o columna a eliminar
* axis: 0 $\Rightarrow$ fila, 1 $\Rightarrow$ columna
* index: equivalente a labels, axis=0 
* columns: equivalente a labels, axis=1

La eliminación entrega un duplicado para guardar el valor duplicado se tiene que sobreescribir el dataFrame

```python
#Obteniendo los índices de la variables
#df['TasaFertilidad'].isna() #=>retorna un vector de verdadero y falso
#df[df['TasaFertilidad'].isna()] #=>retorna los puntos con puntos NA
#El resto obtiene el indice como una lista
indexDrop = df[df['TasaFertilidad'].isna()].index.values.astype(int)[0]

#Botando el valor con la función drop (no sobreescribí el dataframe)
df.drop(index=indexDrop,axis=0)
```

# Modelos de la clase sklearn (estadarización y normalización)
La mayoría de los modelos/funciones de sklearn corresponden a “objetos” de cierta “clase” con propiedades y funciones específicas. En términos prácticos solo se necesita crear una caja mágica que contiene todo lo necesario para aplicar y analizar los modelos.
* from sklearn.xxx import modeloDeseado => Importación del modelo
* from sklearn.xxx import funciónDeseada => Importación de la función
* funciónDeseada() => Aplicación de la función
* nuevoModelo = modeloDeseado() => Creación del modelo
* nuevoModelo.function() => Aplicar alguna función

# standard scaler
StandardScaler es un "modelo" que estandariza los valores entregados retornando la misma matriz aplicando una reducción del valor de la media y dividido por la desviación estandar<br>
StandardScaler(*, with_mean=True, with_std=True)<br>
Parámetros:
* *: set de datos
* with_mean: guarda la media del set de datos de entrenamiento
* with_std: guarda la desviación estándar del set de datos de entrenamiento

Atributos:
* mean_:media de cada variable
* var_: varianza de cada variable
* n_features_in_: Número de variables usadas en el entrenamiento
* feature_names_in_: Nombre de las variables usadas en el entrenamiento

Funciones:
* fit(X): Calcula la media y varianza del set de datos X.
* transform(X): Aplica la transformación al set de datos X.

```python
from sklearn.preprocessing import StandardScaler
escalador=StandardScaler() #Creando el nuevo modelo
escalador.fit(df.iloc[:,3:5]) #ajustando el modelo (da error con variables no numéricas)
print("Media datos originales:",escalador.mean_)
print("Varianza datos originales:",escalador.var_)
nuevosDatos = escalador.transform(df.iloc[:,3:5]) #retorna un arreglo de numpy

print("Media nuevos datos:",np.mean(nuevosDatos,axis=0))
print("Varianza nuevos datos:",np.var(nuevosDatos,axis=0))

tempDF=pd.DataFrame(nuevosDatos,columns=["X1","X2"])
print(ggplot(tempDF)+aes(x="X1",y="X2")+geom_point())
```

# MinMaxScaler
MinMaxScaler es un "modelo" que normaliza los valores dejando cada variable entre el rango 0 y 1<br>
MinMaxScaler(feature_range=(0, 1), *)<br>
Parámetros:
* feature_range: rango de salida para cada variable
* *: set de datos

Atributos:
* data_min_: Mínimo valor para cada variable
* data_max_: Máximo valor para cada variable
* n_features_in_: Número de variables usadas en el entrenamiento
* feature_names_in_: Nombre de las variables usadas en el entrenamiento

Funciones:
* fit(X): Calcula el valor mínimo y máximo del set de datos X.
* transform(X): Aplica la transfo|rmación al set de datos X.

```python
from sklearn.preprocessing import MinMaxScaler
escalador=MinMaxScaler() #Creando el nuevo modelo
escalador.fit(df.iloc[:,3:5]) #ajustando el modelo (da error con variables no numéricas)
print("Valores mínimos originales:",escalador.data_min_)
print("Varianza máximos originales:",escalador.data_max_)
nuevosDatos = escalador.transform(df.iloc[:,3:5]) #retorna un arreglo de numpy

tempDF=pd.DataFrame(nuevosDatos,columns=["X1","X2"])
print(ggplot(tempDF)+aes(x="X1",y="X2")+geom_point())

```

# Sampleo de datos utils.resample
resample es una función que permite realizar los 3 tipos de sampleo mostrados en clases. <br>
resample(*arrays, replace=True, n_samples=None, random_state=None, stratify=None)

Parámetros:
* *arrays: uno o varios arrays de los cuales se realizarán los sampleos.
* replace: True $\Rightarrow$ sampleo con reemplazo, False $\Rightarrow$ sampleo sin reemplazo.
* n_samples: Número de datos a samplear.
* random_state: semilla que permite la replicación del proceso.
* stratify: arreglo para con etiquetas para samplear propocional al número de muestras.

```python
from sklearn.utils import resample
resample(df,n_samples=20,replace=True,random_state=0) #Kiribati se repite
resample(df,n_samples=20,replace=False,random_state=0) #Kiribati ya no se repite, 5 North America
resample(df,n_samples=20,replace=False,random_state=0,stratify=df["Continente"])  #3 North America
```

# OneHotEncoder
One hot encoding transforma una variable de x valores en x variables binarias (una por valor) o x-1 variables binarias (la ausencia de todas las variables indican la presencia de un valor). <br>
La clase OneHotEncoder del módulo sklearn.preprocessing permite realizar este proceso en forma rápida
OneHotEncoder(drop=None,sparse=True)
Parámetro
* drop: método a utilizar para descartar uno de los valores.<br>
None: mantiene todos los valores.<br>
‘first’: bota el primer valor de una variable.<br>
‘if_binary’: bota el primer valor de las variable binarias, y mantiene todos los valores en las otras variables.
* sparse: valor booleano indicando si se quiere devolver una matriz sparse (menos memoria, mala visualización)

<br>Atributos
* categories_: un arreglo con la lista de valores para cada categoría.
* drop_idx_: arreglo indicando el tipo de drop realizado para esa variable.

<br>métodos/funciones
* fit(X): "Entrenamiento" del modelo.
* fit_transform(X): aprende las variables y aplica la transformación.
* transform(X): transforma el set de datos X.
* inverse_transform(X): aplica la transformación inversa a los datos adquiridos.

```python
from sklearn.preprocessing import OneHotEncoder

#Codificador con drop = None
codificador = OneHotEncoder(sparse=False)
print("Continentes:",np.sort(df["Continente"].unique()))
print(df.head(4))
print("Codificador: NONE")
codificador.fit(df["Continente"].values.reshape(-1,1))
print(codificador.transform(df["Continente"].values.reshape(-1,1))[0:4,:])
#Reshape es necesario, no sabe si una variable con 189 con distintos valores o 189 variables distintas

#Codificador con drop = None
codificador = OneHotEncoder(drop='first',sparse=False)
print("Codificador: first")
codificador.fit(df["Continente"].values.reshape(-1,1))
print(codificador.transform(df["Continente"].values.reshape(-1,1))[0:4,:])

```

# OrdinalEncoder (usar con precaución)
Ordinal enconding transforma una variable de x valores en 1 variables binaria con valores de 0 a n-1 donde n es el número de posibles valores de la variable. <br>
La clase OrdinalEncoder del módulo sklearn.preprocessing permite realizar este proceso en forma rápida
OrdinalEncoder(categories='auto')
Parámetro
* categories: 'auto' selecciona en foram "automática" (alfabética) el orden, \[\[arreglo con los valores originales de la variable, donde su orden corresponde a la transformación\]\]

<br>Atributos
* categories_list: Las categorias de las variables encontradas durante el fit.
* n_features_in_: Número de variables utilizadas

<br>métodos/funciones
* fit(X): "Entrenamiento" del modelo.
* fit_transform(X): aprende las variables y aplica la transformación.
* transform(X): transforma el set de datos X.
* inverse_transform(X): aplica la transformación inversa a los datos adquiridos.

```python
from sklearn.preprocessing import OrdinalEncoder

#Codificador con categories = 'auto'
codificador = OrdinalEncoder()
print("Continentes:",np.sort(df["Continente"].unique()))
print(df.head(4))
print("Codificador: auto")
codificador.fit(df["Continente"].values.reshape(-1,1))
print(codificador.transform(df["Continente"].values.reshape(-1,1))[0:4,:])
print("Orden de categorias:",codificador.categories_)
#Reshape es necesario, no sabe si una variable con 189 con distintos valores o 189 variables distintas

#Codificador con drop = None
codificador = OrdinalEncoder(categories=[df["Continente"].unique()])
print("Codificador: first")
codificador.fit(df["Continente"].values.reshape(-1,1))
print(codificador.transform(df["Continente"].values.reshape(-1,1))[0:4,:])
print("Orden de categorias:",codificador.categories_)
```

# Transformación 1:1
Dentro de las transformaciones 1:1 existe la función power_transform de sklearn.preprocessing que permite dos tipos de transformaciones yeo-johnson y box-cox
power_transform(X, method='yeo-johnson',standardize=True)<br>
Parámetro
* method: transformación a aplicar
'yeo-johnson': permite valores en cualquier rango de valores.<br>
‘box-cox’: permite solo valores positivos
* standardize: Que la transformación de los datos sea N(0,1)

Recuerden que como es una función solo retorna la matriz con nuevos valores

```python
from sklearn.preprocessing import power_transform

print(ggplot(df)+aes(x="TasaFertilidad")+geom_density()+labs(title="Original"))

tempData=power_transform(df["TasaFertilidad"].values.reshape(-1,1),method="box-cox")
tempDF=pd.DataFrame(tempData,columns=["newTasaFertilidad"])
print(ggplot(tempDF)+aes(x="newTasaFertilidad")+geom_density()+labs(title="box-cox"))
```

# Transformaciones Binning 
La función qcut de pandas permite realizar la mayoría de los binning.<br>
qcut(x, q, labels=None)<br>
Parámetros
* x: arreglo de una dimensión o panda series
* q: número de bins (todos los bins con la misma cantidad), o vector con quantiles de la separación
* labels: OPCIONAL, etiquetas a devolver para los puntos que caen dentro de cada separación

Atención, esta función no funciona para variables categóricas.

```python
#Separando la tasa de fertilidad en q espacios con el mismo número de puntos
print(pd.qcut(df["TasaFertilidad"],q=3))

#Separando la tasa de fertilidad en quantiles definidos por uno
print("\n\n",pd.qcut(df["TasaFertilidad"],q=[0,0.2,0.5,0.8,1.0]))

#Separando la tasa de fertilidad en quantiles definidos por uno y sus respectivas etiquetas
print("\n\n",pd.qcut(df["TasaFertilidad"],q=[0,0.2,0.5,0.8,1.0],labels=[1.4,1.8,2.7,5.5]))

```

# PCA
La clase PCA de sklearn.decomposition de pandas permite aplicar PCA.<br>
PCA(n_components=None, *)<br>

Parámetros:
* n_components: Número de componentes a calcular en caso que no se quiera calcular todos los componente (el número de variables)
* *: Matriz con los datos

Atributos:
* components_: Matriz con las componentes ("para realizar las transformaciones")
* explained_variance_: canitdad de varianza explicada por cada componente
* explained_variance_ratio_: Porcentaje de varianza explicada (la suma debería dar 1 si n_components no es definido)

Funciones:
* fit(X): Entrenar el modelo con los datos de X
* transform(X): Transofrmar los datos X aplicando la conversión aprendida con fit

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df.iloc[:,3:5])
tempData = pca.transform(df.iloc[:,3:5])
tempData = pd.DataFrame(tempData,columns=["PC1","PC2"])
tempData["Continente"] = df["Continente"]
print("Varianza explicada:",pca.explained_variance_ratio_)
print(ggplot(df)+aes(x="TasaFertilidad",y="IngresoPerCapita",color="Continente")+geom_point()+theme_bw())
print(ggplot(tempData)+aes(x="PC1",y="PC2",color="Continente")+geom_point()+theme_bw())
```

```python
#Replicando el proceso estandarizando los datos
from sklearn.preprocessing import StandardScaler
escalador=StandardScaler() #Creando el nuevo modelo
escalador.fit(df.iloc[:,3:5]) #ajustando el modelo (da error con variables no numéricas)
nuevosDatos = escalador.transform(df.iloc[:,3:5]) #retorna un arreglo de numpy

from sklearn.decomposition import PCA
pca = PCA()
pca.fit(nuevosDatos)
tempData = pca.transform(nuevosDatos)
tempData = pd.DataFrame(tempData,columns=["PC1","PC2"])
tempData["Continente"] = df["Continente"]
print("Varianza explicada:",pca.explained_variance_ratio_)
print(ggplot(df)+aes(x="TasaFertilidad",y="IngresoPerCapita",color="Continente")+geom_point()+theme_bw())
print(ggplot(tempData)+aes(x="PC1",y="PC2",color="Continente")+geom_point()+theme_bw())
```

# Kernel PCA
La clase kernelPCA de sklearn.decomposition de pandas permite aplicar kernel PCA.<br>
KernelPCA(n_components=None, *, kernel='linear', gamma=None, degree=3, coef0=1)<br>

Parámetros:
* n_components: Número de componentes a calcular en caso que no se quiera calcular todos los componente (el número de variables o puntos -1, lo que sea mayor)
* *: Matriz con los datos.
* kernel: Selección del tipo de kernel (‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘cosine’, ‘precomputed’)
* gamma, degree, coef0: hyperparameters of the different kernels.

Atributos:
* 

Funciones:
* fit(X): Entrenar el modelo con los datos de X
* transform(X): Transofrmar los datos X aplicando la conversión aprendida con fit

Atención, tal como con Kernel PCA, la estandarización si afecta el proceso

```python
from sklearn.decomposition import KernelPCA

#Replicando el proceso estandarizando los datos
from sklearn.preprocessing import StandardScaler
escalador=StandardScaler() #Creando el nuevo modelo
escalador.fit(df.iloc[:,3:5]) #ajustando el modelo (da error con variables no numéricas)
print("Media datos originales:",escalador.mean_)
print("Varianza datos originales:",escalador.var_)
nuevosDatos = escalador.transform(df.iloc[:,3:5]) #retorna un arreglo de numpy


pca = KernelPCA(n_components=2,kernel="rbf")
pca.fit(nuevosDatos)
tempData = pca.transform(nuevosDatos)
tempData = pd.DataFrame(tempData,columns=["PC1","PC2"])
tempData["Continente"] = df["Continente"]
print(ggplot(df)+aes(x="TasaFertilidad",y="IngresoPerCapita",color="Continente")+geom_point()+theme_bw())
print(ggplot(tempData)+aes(x="PC1",y="PC2",color="Continente")+geom_point()+theme_bw())
```

# ICA

FastICA(n_components=None, *)<br>

Parámetros:

* *: Matriz con los datos.

Atributos:

Funciones:

```python
from sklearn.decomposition import FastICA

#Replicando el proceso estandarizando los datos
from sklearn.preprocessing import StandardScaler
escalador=StandardScaler() #Creando el nuevo modelo
escalador.fit(df.iloc[:,3:5]) #ajustando el modelo (da error con variables no numéricas)
print("Media datos originales:",escalador.mean_)
print("Varianza datos originales:",escalador.var_)
nuevosDatos = escalador.transform(df.iloc[:,3:5]) #retorna un arreglo de numpy


ica = FastICA()
ica.fit(df.iloc[:,3:5])
tempData = ica.transform(df.iloc[:,3:5])
tempData = pd.DataFrame(tempData,columns=["PC1","PC2"])
tempData["Continente"] = df["Continente"]
print(ggplot(df)+aes(x="TasaFertilidad",y="IngresoPerCapita",color="Continente")+geom_point()+theme_bw())
print(ggplot(tempData)+aes(x="PC1",y="PC2",color="Continente")+geom_point()+theme_bw())

ica = FastICA()
ica.fit(nuevosDatos)
tempData = ica.transform(nuevosDatos)
tempData = pd.DataFrame(tempData,columns=["PC1","PC2"])
tempData["Continente"] = df["Continente"]
print(ggplot(tempData)+aes(x="PC1",y="PC2",color="Continente")+geom_point()+theme_bw())
```

# Variance Threshold 
La clase VarianceThreshold de skelearn.feature_selection es una búsqueda de "tipo" ranking basada exclusivamente en la varianza de los valores.<br>
VarianceThreshold(threshold=0.0)<br>

Parámetros:
* threshold: Cualquier variable con varianza menor a threshold es eliminada.

Atributos:
* variances_: Varianza de cada componente
* n_features_: Número de características usadas para el entrenamiento

Funciones:
* fit(X): Entrenar el modelo con los datos de X
* transform(X): Seleccionar las variables de X aplicando el threshold seleccionado

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(0.1) #cambiar valor a 10
selector.fit(df.iloc[:,3:5])
print(selector.variances_)
print(selector.transform(df.iloc[:,3:5]))

#Si no se realiza el fit, el modelo no puede aplicar transform. Una solución es la función fit_transform
#print(selector.fit_transform(df.iloc[:,3:5]))
```
